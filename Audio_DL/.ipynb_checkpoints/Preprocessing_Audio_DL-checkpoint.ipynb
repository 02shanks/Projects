```python
import librosa, librosa.display
import matplotlib.pyplot as plt
import numpy as np
from pathlib import Path
```

```python
#Waveform
signal, sr = librosa.load('WAV_1MG.wav', sr=22050) ## sr*T --> 22050*33(here)
```

```python
librosa.display.waveshow(signal, sr=sr)
plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.show()
```

```python
## fft --> spectrum
fft = np.fft.fft(signal)
magnitude = np.abs(fft)
frequency = np.linspace(0, sr, len(magnitude))


plt.plot(frequency, magnitude)
plt.xlabel("Frequency")
plt.ylabel("Magnitude")
plt.show()
```

```python
magnitude.shape
```

```python
## As plot is symmertric we need just half part to extract features
fft = np.fft.fft(signal)
magnitude = np.abs(fft)
frequency = np.linspace(0, sr, len(magnitude))

left_frequency = frequency[:int(len(frequency)/2)]
left_magnitude = magnitude[:int(len(magnitude)/2)]



plt.plot(left_frequency, left_magnitude)
plt.xlabel("Frequency")
plt.ylabel("Magnitude")
plt.show()
```

```python
## Now need to capture time variable
## stft --> spectrogram

hop_length = 512 # len of window u want to capture from fft graph
n_fft = 2048 # no of sample in that window

stft = librosa.core.stft(signal, hop_length= hop_length, n_fft=n_fft)
spectrogram = np.abs(stft)

librosa.display.specshow(spectrogram, sr=sr , hop_length=hop_length)
plt.xlabel("Time")
plt.ylabel("Frequency")
plt.colorbar()
plt.show()
```

```python
log_spectrogram = librosa.amplitude_to_db(spectrogram)
librosa.display.specshow(log_spectrogram, sr=sr , hop_length=hop_length)
plt.xlabel("Time")
plt.ylabel("Frequency")
plt.colorbar()
plt.show()
```
